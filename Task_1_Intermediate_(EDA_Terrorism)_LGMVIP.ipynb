 **DATA VISUALIZATION**

**TERRORIST ATTACK IN EACH YEAR**

```python
plt.subplots(figsize=(15,6))
sns.countplot(x='year',data=df,color='red')
plt.xticks(rotation=45,fontsize=10)
plt.ylabel('Number of Attacks Each Year',fontsize=10)
plt.title('Terrorist Attack Each Year',fontsize=15)
plt.show
```

**REGION-WISE ATTACKS**

```python
pd.crosstab(df.year, df.Region).plot(kind='area',stacked=False,figsize=(15,6))
plt.title('Terrorist Activities By Region In Each Year',fontsize=15)
plt.ylabel('Number of Attacks',fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(10,10))
sns.barplot(df['Region'].value_counts()[:10].index,df['Region'].value_counts()[:10].values,color='darkred')
plt.title('Top Region Affected')
plt.xlabel('Region')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

**COUNTRIES WITH HIGHEST NUMBER OF ATTACKS**

```python
plt.subplots(figsize=(15,6))
sns.barplot(df['Country'].value_counts()[:15].index,df['Country'].value_counts()[:15].values,color='blue')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 90)
plt.show()
```

**ATTACK ON CITIES**

```python
df['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(15,6),color='orange')
plt.xticks(rotation = 90)
plt.xlabel("City",fontsize=10)
plt.ylabel("Number of attack",fontsize=10)
plt.title("Top 10 most effected city",fontsize=15)
plt.show()
```

**ACTIVE TERRORIST GROUPS**

```python
df['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='green',figsize=(15,6))
plt.title("Top 10 terrorist group attack",fontsize=15)
plt.xlabel("terrorist group name",fontsize=10)
plt.ylabel("Attack number",fontsize=10)
plt.show()
```

**TYPES OF ATTACKS**

```python
df['Attacktype'].value_counts().plot(kind='bar',figsize=(15,6),color='grey')
plt.xticks(rotation = 90)
plt.title("Name of attacktype",fontsize=15)
plt.xlabel("Attacktype",fontsize=10)
plt.ylabel("Number of attack",fontsize=10)
plt.show()
```

**TYPE OF TARGET**

```python
df['Targettype'].value_counts().plot(kind='bar',figsize=(15,6),color='black')
plt.xticks(rotation = 90)
plt.title("Types of Targets",fontsize=15)
plt.xlabel("Target Types",fontsize=10)
plt.ylabel("Number of attack",fontsize=10)
plt.show()
```
